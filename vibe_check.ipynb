```python
import pandas as pd
from gensim.utils import simple_preprocess
from transformers import pipeline
```

```python
# load yelp dataset

url = 'https://www.dropbox.com/s/xds4lua69b7okw8/yelp.csv?dl=1'
data = pd.read_csv(url)
print(data)
```

```python
data.describe()
```

```python
# only take extreme stars data
data = data[data['stars'].isin([1,5])]
print(data)
```

```python
# initialize sentiment analysis pipeline

sentiment_analysis = pipeline('sentiment-analysis')
```

```python
# run sentiment analysis on a few examples

examples = data['text'].sample(5).tolist()
results = sentiment_analysis(examples)

for i, text in enumerate(examples):
  print(f'THIS IS THE TEXT:\n {text} \nTHIS IS THE SENTIMENT: {results[i]}\n')
  print('-----------------oOo----------------')
```

```python
# setting up flask api
```

```python
%%writefile sentiment_analysis.py

from flask import Flask, request, jsonify
from functools import lru_cache

from transformers import pipeline

app = Flask(__name__)

# initialize sentiment analysis pipeline

sentiment_analysis = pipeline('sentiment-analysis')

# cache using dict
cache = {}

@app.route('/sentiment-analysis', methods=['POST'])
def get_sentiment():
  data = request.get_json()
  text = data.get('text')

  # check cache:
  if text in cache:
    result = cache[text]
  else:
    result = sentiment_analysis(text)
    cache[text] = result
  return jsonify(result)

if __name__ == '__main__':
  app.run(host='0.0.0.0', port=5000)
```
